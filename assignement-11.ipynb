```python
Q3. Which Python statements are used to catch and handle exceptions? Explain with an 
example.

Ans3. To handle exception in python, we use try-except block. The advantage of using this
block is that even if the exception occurs in that case our whole program does not terminate
abruptly. The rest of the program is executed normally.
```

```python
try:
    a=10
    b=a/0
except Exception as e:
    print(e)
print("This statement is going to be printed since exception is handled successfully")
print("exception handling avoid the abrupt termination of the program")
```

```python
Q5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example.
Ans5. In Python, custom exceptions are user-defined exceptions that we can create by subclassing 
the built-in Exception class. By creating custom exceptions, 
we can handle specific error cases more effectively and provide more meaningful error 
messages, tailored to the context of the application.
```

```python
class numexception(Exception):
    def __init__(self,msg):
        self.msg=msg
    def num_check(num):
        if num>=1000:
            raise numexception("The number should be less than 1000")
        else:
            print("valid number")

try:
    num=int(input("enter any number less than 1000"))
    numexception.num_check(num)
except numexception as e:
    print(e)    
```

```python
class validateageexception(Exception):
    def __init__(self,msg):
        self.msg=msg
    def validage(age):
        if age<0:
            raise validateageexception("age is negative")
        elif age>120:
            raise validateageexception("age is very very high")
        else:
            print("age is valid")

try:
    age=int(input("pleae enter your age"))
    validateageexception.validage(age)
except validateageexception as e:
    print(e)
```
