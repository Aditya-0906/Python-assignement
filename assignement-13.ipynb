```python
Q2. Why threading module used? Write the use of the following functions:

1. activeCount()
2. currentThread()
3. enumerate()

Ans2: The threading module in Python is used to provide a way to run multiple threads (smaller units of a process) concurrently. 
It is part of Python standard library and offers a high-level interface to create and manage threads. The threading module is 
essential for building concurrent applications in Python that require running multiple tasks at the same time, especially for 
I/O-bound tasks, improving the efficiency, responsiveness, and design of our programs.

1. active_count():  The activeCount() function returns the number of Thread objects currently alive in the program. It counts 
how many threads are running, including the main thread.

```

```python
import threading

def task(id):
    print("Thread %d is running"% id)

thread=[threading.Thread(target=task,args=(i,)) for i in range(3)]
for t in thread:
    t.start()
print("The number of active threads are ",threading.active_count())
```

```python
2. currentThread():  The currentThread() function returns the Thread object corresponding to the currently executing thread.
It helps in getting the current thread's identity, which can be useful for debugging or logging. This function is useful when you 
want to identify the thread that is executing some code, especially in multithreaded environments where multiple threads are working 
simultaneously.
```

```python
import threading

def thread_1(i):
    print("active thread right now is",threading.current_thread())
    print("value by Thread 1:",i)

def thread_2(i):
    print("active thread right now is",threading.current_thread())
    print("value by Thread 2:",i)

def thread_3(i):
    print("active thread right now is", threading.current_thread())
    print("value by Thread 3:",i)

t1=threading.Thread(target=thread_1,args=(1,))
t2=threading.Thread(target=thread_2,args=(2,))
t3=threading.Thread(target=thread_3,args=(3,))

print("active thread right now is ",threading.current_thread())

t1.start()
t2.start()
t3.start()
```

```python
3. enumerate(): The enumerate() function returns a list of all the Thread objects currently alive. This includes both the main thread 
and any other threads that are still running.

```

```python
import threading

def task():
    pass

thread=[threading.Thread(target=task) for i in range(3)]

for t in thread:
    t.start()

for thread in threading.enumerate():
    print(thread.name)
```

```python
Q3. Explain the following functions:
1. run()
2. start()
3. join()
4. isAlive()

Ans3. The run() method is used to define the entry point of the thread. When you create a thread by subclassing 
Thread and overriding its run() method, this method defines the code that the thread will execute. Typically, we
don’t call run() directly. Instead, we use start(), which internally calls run(). However, if we subclass Thread, 
we can override run() to define the specific task we want the thread to perform.
```

```python
import threading

class mythread(threading.Thread):
    def run(self):
        print("the execution of program will start from here")

t=mythread()
t.start()
```

```python
2. start(): The start() method is used to initiate the threads execution. When start() is called, it creates a new thread of 
execution and immediately begins running the run() method (or the target function). 

3. join():  The join() method is used to make the main program (or the calling thread) wait for the thread to finish its execution. 
It blocks the calling thread until the thread whose join() is called has completed.

4. isAlive(): The isAlive() method was used to check whether a thread is still running (alive). It returns True if the thread is 
still alive (i.e., it has been started but has not yet completed), and False if the thread has finished its task or hasn’t been 
started yet.

Now lets understand all the three start(), join() and isAlive() with the help of a program.
```

```python
import threading
import time

def task():
    time.sleep(5)
    print("The thread is complete")

t=threading.Thread(target=task)
t.start()

print("Checking if the thread is alive ",t.is_alive())
t.join()

print("Is the thread still alive after join ",t.is_alive())
```

```python
Q6. Explain deadlocks and race conditions.

Ans6. Deadlock: A deadlock is a situation in concurrent programming where two or more threads (or processes) are blocked forever, 
waiting for each other to release resources that they need to continue their execution. This leads to a situation where no thread 
can proceed, effectively freezing the program or system.

How Deadlock Occurs:
Deadlock typically occurs when the following four conditions are met (often referred to as the Coffman conditions):

Mutual Exclusion: Only one thread can access a resource at a time. If one thread holds a resource, others must wait for it to 
be released.

Hold and Wait: A thread that holds at least one resource is waiting for additional resources that are currently held by other threads.

No Preemption: Resources cannot be forcibly taken from threads holding them. They must be released voluntarily.

Circular Wait: A set of threads are waiting for resources held by the next thread in a circular chain. For example, Thread A is 
waiting for a resource held by Thread B, and Thread B is waiting for a resource held by Thread A.
```

```python
Race condition: A race condition occurs when multiple threads or processes attempt to read and write shared data at the same time, 
leading to unpredictable behavior. The final outcome depends on the timing or sequence of thread execution, which can result in 
inconsistent or erroneous behavior.

Race conditions typically happen when:

1. Two or more threads are accessing shared data simultaneously.
2. At least one of them modifies the data.
3. There is no proper synchronization (e.g., locks) to ensure that the data is accessed and modified safely.
```
