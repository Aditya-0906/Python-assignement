```python
Q2. Write a python program to print Python Exception Hierarchy.
```

```python
import inspect

def print_exception_hierarchy(cls, level=0):
    """Recursively prints the exception class hierarchy."""
    print('  ' * level + cls.__name__)
    for subclass in cls.__subclasses__():
        print_exception_hierarchy(subclass, level + 1)

# Starting with the base Exception class
print("Python Exception Hierarchy:")
print_exception_hierarchy(Exception)
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans4. The LookupError class in Python is a base class for exceptions that are raised when
a lookup operation (such as accessing a key in a dictionary or an index in a list) fails. 
It serves as a common parent class for more specific exceptions like KeyError and 
IndexError.
```

```python
#Exmaple for KeyError:
import logging
logging.basicConfig(filename="error.log",level=logging.ERROR)
try:
    dic={"abhi":123,"aman":345,"krishna":908}
    value=dic["anuj"]
except KeyError as e:
    logging.error("Handling keyerror exception {}".format(e))
    print(e)
```

```python
# Example for IndexError:
import logging
logging.basicConfig(filename="error.log",level=logging.ERROR)
try:
    l=[2,3,4]
    print(l[7])
except IndexError as e:
    logging.error("Handline Index error here {}".format(e))
    print(e)
```

```python
Q5. Explain ImportError. What is ModuleNotFoundError?

Ans5. ImportError is an exception that is raised when an imported module or a specific 
object from a module cannot be found or loaded.

ModuleNotFoundError is a more specific subclass of ImportError that is raised when a 
module is not found during an import statement. This error is typically seen when the 
module you are trying to import does not exist in the current Python environment.

It was introduced in Python 3.6 to provide more clarity when dealing with module imports.
Before Python 3.6, ImportError was used to handle all import-related errors, 
including the case where a module could not be found. In Python 3.6 and later, 
ModuleNotFoundError is raised when Python is unable to locate a module.
```

```python
Q6. List down some best practices for exception handling in python.
Ans6. The best practices for exception handling in python are as follows:

1. Always use a specific Exception: It is a good practise in python to use the specific 
class for handling an exception. For example: if we are handling a ZeroDivisionError then
can handle it with Exception or ArithmeticError class as well, however it is best to 
handle it with ZeroDivisionError only.

2. Always print a proper message in case exception occurs: It is always better to provide descriptive reason for the occurance of the exception,
so that if any other person also working on the project, so they can identify easily what went 
wrong.

3. Always try to log your error: This point is very similar to our previous point. Here,
we need to log the error so that other persons working on the project may figure out
the errors that occured while running the program. So that they can improvise the code easily.

4. Always try to avoid multiple exception handling: Instead of using mulitple except blocks
(or multiple exception classes), we need to use a single specific sub-class (not super class Exception)
in order to handle the exception.

5. Document all the errors: In an industry grade program, it is mandator that we document
all the error and purpose of the modules that we made, so that our program or module 
can be easily understood and improvised in the future.

6. Clean up all the resources: While performing exception handling or writing the code,
it is required that we relase all the resources used by the module while execution, once
the execution of module is finished.
For ex: When we open the file then it is mandatory to close it. So we generally close the
file in finally block so that the file can be closed in any case (even if the exception 
occurs).                                                                             
                                                                             
```
