```python
Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode
of file opening.

Ans1. The function used to open a file in python is open("file_path","open_mode"). The parametres passed to this function are file name and open mode.
The different modes for opening a file are as follows:

1. 'f'--> open for reading.
2. 'w'--> open for writing, truncating for the file first.
3. 'x'--> create a new file and open it for writing.
4. 'a'--> oprn for writing, appending to the end of the file if it exists.
5. 'b'--> binary mode
6. 't'--> text mode(default)
7. '+'--> open a disk file for updating (reading and writing).
```

```python
Q2. Why close() function is used? Why is it important to close a file?
Ans2. In python close() function is used to close the file and release the memory occupied by the file variable. It is important to close a file 
because of following reasons:

1. When we open a file, the operating system allocates resources to manage the file access. Closing the file ensures that these resources are 
released and made available for other processes.
2. When we write data to a file, the data is often first buffered in memory. The close() function ensures that all the data in the buffer 
is properly written to the file before closing it.
3. Not closing a file properly can sometimes result in file corruption, especially in cases of writing operations
```

```python
Q4. Explain the following with python code: read(), readline() and readlines().
Ans4. All the three methods are used to read the content of a file, however there are differences in how these methods work and how these methods 
provide the content of the file.

1. read() reads the entire file content as one big string.
2. readline() reads one line at a time from the file.
3. readlines() returns a list where each element is one line from the file.
```

```python
# Program for read()

file=open("test1.txt",'w')
file.write("Hello!\n")
file.write("This program is to explain the difference amonng read(), readline() and readlines()\n")
file.write("Thank you\n")
file.close()

file1= open("test1.txt",'r')
print(file1.read())
file1.close()
```

```python
# Program for realine()

file=open("test1.txt",'r')
line1= file.readline()
print(line1)

line2=file.readline()
print(line2)

line3=file.readline()
print(line3)

file.close()
```

```python
# Program for readlines()

file=open("test1.txt",'r')
lines=file.readlines()

for i in lines:
    print(i)
```

```python
Q5. Explain why with statement is used with open(). What is the advantage of using with statement and
open() together?

Ans5. Python provides a more efficient and cleaner way to manage files using the with statement. This method ensures 
that the file is automatically closed once the block of code inside the with statement is finished, even if an exception occurs. So we do not
need to manually call the close() once we have opened the file with "with open()".
```

```python
Q6. Explain the write() and writelines() functions. Give a suitable example.

Ans6. Both the functions are used to add data in the file. However the difference in these functions is about how they put data inside the file.
Like write() puts a string inside the file, whereas writelines() is used to put iterables in the file like list or tuple of strings. With the write()
we enter strings in a file one by one whereas with the writelines(), we can make a list or tuple of string and enter all the strings at once.
```

```python
# Write function:

file=open("test2.txt",'w')
file.write("This is my first line\n")
file.write("This is my second line\n")
file.write("This is my third line\n")
file.close()
```
